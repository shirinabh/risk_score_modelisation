```python
def _clean_cell(x):
    if pd.isna(x):
        return pd.NA
    s = str(x).strip()

    if s == "" or s.lower() in {"nan", "none", "null"}:
        return pd.NA
    return s

business_failures_codes_df["period_scope"] = business_failures_codes_df["period_scope"].map(_clean_cell)
```

```python
import pandas as pd
import numpy as np

REG = "11"   # Île-de-France
A21 = "L"    # Activités immobilières

# --- total d'entreprises INSEE (annuel) ---
base = companies_df[(companies_df["REG"]==REG) & (companies_df["A21"]==A21)].copy()
base["ANNEE"] = pd.to_numeric(base["ANNEE"], errors="coerce")
base = base[base["ANNEE"].between(2014, 2021)]
base["NB_UL"] = pd.to_numeric(base["NB_UL"], errors="coerce")

# --- défaillances BdF (cumul 12 mois) -> prendre la dernière valeur de chaque année ---
x = df_secteur[["date","nb_defaillances"]].copy()
x["date"] = pd.to_datetime(x["date"], errors="coerce")
x["ANNEE"] = x["date"].dt.year
x["MOIS"]  = x["date"].dt.month
x["nb_defaillances"] = pd.to_numeric(x["nb_defaillances"], errors="coerce")

# on garde, pour chaque année, la valeur du mois max (idéalement 12)
last_of_year = (x.dropna(subset=["ANNEE","nb_defaillances"])
                  .sort_values(["ANNEE","date"])
                  .groupby("ANNEE", as_index=False)
                  .tail(1)                         # dernière date de l'année
                  .loc[:, ["ANNEE","nb_defaillances"]]
                  .rename(columns={"nb_defaillances":"defaillances_annee"}))

# --- jointure + taux ---
z = (last_of_year.merge(base[["ANNEE","NB_UL"]], on="ANNEE", how="inner")
                .sort_values("ANNEE"))
z = z[(z["NB_UL"]>0) & z["defaillances_annee"].notna()].copy()
z["taux"] = z["defaillances_annee"] / z["NB_UL"]

# --- split apprentissage / validation ---
train = z[z["ANNEE"].between(2014, 2019)].copy()
valid = z[z["ANNEE"].between(2020, 2021)].copy()

# --- score percentile (ECDF) borné ---
vals = np.sort(train["taux"].dropna().values)
def ecdf_score(v):
    if not len(vals) or pd.isna(v): 
        return np.nan
    p = np.searchsorted(vals, v, side="right") / len(vals)  # 0..1
    p = float(np.clip(p, 0.01, 0.99))
    return round(100*p, 2)

z["risk_score"] = z["taux"].apply(ecdf_score)

print(z)
print("train:", train["ANNEE"].tolist())
print("valid:", valid["ANNEE"].tolist())
print("taux min/max (train):", float(train["taux"].min()), float(train["taux"].max()))
```

```python
import pandas as pd
import numpy as np
import dataiku
from dataiku import Dataset

# =========================
# Données
# =========================
codes_df = dataiku.Dataset("bdf_business_failures_codes").get_dataframe()
fail_df  = dataiku.Dataset("bdf_business_failures").get_dataframe()
insee_df = dataiku.Dataset("insee_companies").get_dataframe()  # ANNEE, REG, A21, NB_UL

# =========================
# Paramètres
# =========================
regions_keep = [
 'Hauts-de-France','Bretagne','Normandie','Grand Est','Pays de la Loire',
 'Centre-Val de Loire','Bourgogne-Franche-Comté','Nouvelle-Aquitaine',
 'Auvergne-Rhône-Alpes','Occitanie',"Provence-Alpes-Côte d'Azur",
 'Corse','Île-de-France'
]
sectors_keep = [
 'Construction','Transports et entreposage','Hébergement et restauration',
 'Information et communication',"Activités financières et d'assurance",
 'Activités immobilières'
]
reg_code = {
 'Île-de-France':'11','Centre-Val de Loire':'24','Bourgogne-Franche-Comté':'27',
 'Normandie':'28','Hauts-de-France':'32','Grand Est':'44','Pays de la Loire':'52',
 'Bretagne':'53','Nouvelle-Aquitaine':'75','Occitanie':'76','Auvergne-Rhône-Alpes':'84',
 "Provence-Alpes-Côte d'Azur":'93','Corse':'94'
}
a21_code = {
 'Construction':'F','Transports et entreposage':'H','Hébergement et restauration':'I',
 'Information et communication':'J',"Activités financières et d'assurance":'K',
 'Activités immobilières':'L'
}

# =========================
# Prépa BDF
# =========================
fail_df = fail_df.copy()
fail_df["date"] = pd.to_datetime(fail_df["date"])
fail_df["ANNEE"] = fail_df["date"].dt.year
fail_df = fail_df[(fail_df["ANNEE"]>=2014) & (fail_df["ANNEE"]<=2021)]

codes_trim = codes_df[
    (codes_df["period_scope"]=="trimestre") &
    (codes_df["region_name"].isin(regions_keep)) &
    (codes_df["sector_name"].isin(sectors_keep))
].copy()

out = []
for _, r in codes_trim.iterrows():
    col = r["business_failure_code_bq"]
    if col not in fail_df.columns:
        continue
    tmp = fail_df.groupby("ANNEE", as_index=False)[col].sum(numeric_only=True)
    tmp.rename(columns={col:"defaillances_annee"}, inplace=True)
    tmp["region"] = r["region_name"]
    tmp["sector"] = r["sector_name"]
    out.append(tmp)

if not out:
    raise RuntimeError("Aucun couple région/secteur valide (colonnes absentes).")

bdf_year = pd.concat(out, ignore_index=True)

# Ajout REG / A21
bdf_year["REG"] = bdf_year["region"].map(reg_code)
bdf_year["A21"] = bdf_year["sector"].map(a21_code)

# =========================
# INSEE: types + plage
# =========================
insee_df = insee_df.copy()
insee_df["ANNEE"] = pd.to_numeric(insee_df["ANNEE"], errors="coerce").astype("Int64")
insee_df = insee_df[(insee_df["ANNEE"]>=2014) & (insee_df["ANNEE"]<=2021)]
insee_df["NB_UL"] = pd.to_numeric(insee_df["NB_UL"], errors="coerce")
insee_df["REG"] = insee_df["REG"].astype(str)
insee_df["A21"] = insee_df["A21"].astype(str)

# =========================
# Jointure + taux
# =========================
z = bdf_year.merge(insee_df[["ANNEE","REG","A21","NB_UL"]],
                   on=["ANNEE","REG","A21"], how="left")
z = z.dropna(subset=["NB_UL","defaillances_annee"])
z["taux"] = z["defaillances_annee"] / z["NB_UL"]
z = z[["ANNEE","defaillances_annee","NB_UL","region","sector","taux"]] \
      .sort_values(["region","sector","ANNEE"]).reset_index(drop=True)

print(z.head(20))
print(z.groupby(["region","sector"]).size().unstack(fill_value=0).head())

# =========================
# Export Dataiku (optionnel)
# =========================
try:
    dataiku.Dataset("annual_default_rates").write_with_schema(z)
    print("Écrit -> annual_default_rates")
except Exception as e:
    print("Pas d'export Dataiku :", e)

# ==========================================================
# >>> PUSH VERS SNOWFLAKE : DATALAKE.PUBLIC.ANNUAL_DEFAULT_RATES <<<
#   - Connexion Dataiku : snowflake_con
#   - Database : DATALAKE
#   - Schema   : PUBLIC
#   - Table    : ANNUAL_DEFAULT_RATES
#   - Dataset  : sf_annual_default_rates (créé si absent)
# ==========================================================

SF_CONN   = "snowflake_con"
SF_DB     = "DATALAKE"
SF_SCHEMA = "PUBLIC"
SF_TABLE  = "ANNUAL_DEFAULT_RATES"
DS_NAME   = "annual_default_rates_sf"

# -- Création robuste du dataset via create_dataset (type SQL_TABLE) --
client = dataiku.api_client()
project = client.get_project(dataiku.default_project_key())

# si un dataset du même nom existe partiellement/corrompu, on le supprime avant (sécurisé)
try:
    _tmp = project.get_dataset(DS_NAME)
    # il existe bien -> on ne fait rien
except Exception:
    # créer le dataset
    ds = project.create_dataset(
        name=DS_NAME,
        type="SQL_TABLE",
        params={
            "connection": SF_CONN,
            "table": SF_TABLE,        # non quoté -> ANNUAL_DEFAULT_RATES
            "schema": SF_SCHEMA,      # PUBLIC
            "db": SF_DB,              # << clé attendue ici (pas "catalog")
            "writeMode": "OVERWRITE"  # ou "APPEND"
        },
        formatType=None,
        formatParams=None
    )

# vérif : le dataset est bien visible dans le projet
names = [d["name"] for d in project.list_datasets()]
print("Datasets du projet (extrait) ->", names[:10], ("... total:", len(names)))

if DS_NAME not in names:
    raise RuntimeError(f"Dataset '{DS_NAME}' introuvable après création. Vérifie la connexion '{SF_CONN}' et les droits.")

# normalisation types pour éviter les dtypes pandas nullable
z_for_write = z.copy()
for col in z_for_write.columns:
    if str(z_for_write[col].dtype) == "Int64":
        z_for_write[col] = z_for_write[col].astype("int64")

# écriture
sf_ds = Dataset(DS_NAME)
sf_ds.write_with_schema(z_for_write)
print(f"✅ Écrit dans Snowflake : {SF_DB}.{SF_SCHEMA}.{SF_TABLE} via dataset '{DS_NAME}' et connexion '{SF_CONN}'")
```
